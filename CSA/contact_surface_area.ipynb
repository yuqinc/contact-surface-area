```python
dfu.columns=['index','num','lesion-free']
dfua.columns=['index','num','lesion-free+ack']
dft.columns=['index','num','trans']
dfta.columns=['index','num','trans+ack']
dfc.columns=['index','num','cis']
dfca.columns=['index','num','cis+ack']
```

```python
plt.plot(dfu['num'],dfu['lesion-free'],color='green',label="lesion-free")
plt.plot(dfua['num'],dfua['lesion-free+ack'],color='purple',label="lesion-free+ack")
plt.legend(fontsize=14)
plt.xlim=(0,8000)
plt.ylim=(9,16)
```

```python
xu = dfu['lesion-free'] 
xua = dfua['lesion-free+ack']
xt = dft['trans']
xta = dfta['trans+ack']
xc = dfc['cis']
xca = dfca['cis+ack']
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(8,8),sharex=True, sharey=True)
plt.tight_layout()
fig.subplots_adjust(top=1)


sns.kdeplot(xu, ax=ax1,shade=True, color='green' )
sns.kdeplot(xua, ax=ax1,shade=True,color='purple')

sns.kdeplot(xt, ax=ax2,shade=True,color='green')
sns.kdeplot(xta, ax=ax2,shade=True,color='purple')

sns.kdeplot(xc, ax=ax3,shade=True, color='green')
sns.kdeplot(xca, ax=ax3,shade=True,color='purple')
```
