```python
xu = dfu['lesion-free'] 
xua = dfua['lesion-free+ack']
xt = dft['trans']
xta = dfta['trans+ack']
xc = dfc['cis']
xca = dfca['cis+ack']
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,figsize=(10,12),sharex=True, sharey=True)
sns.despine(left=True)
plt.tight_layout()
fig.subplots_adjust(top=1)


sns.kdeplot(xu, ax=ax1,shade=True, color='green' )
sns.kdeplot(xua, ax=ax1,shade=True,color='purple')

sns.kdeplot(xt, ax=ax2,shade=True,color='green')
sns.kdeplot(xta, ax=ax2,shade=True,color='purple')

sns.kdeplot(xc, ax=ax3,shade=True, color='green')
sns.kdeplot(xca, ax=ax3,shade=True,color='purple')

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
fig.savefig('csa.png', dpi=600)
```

```python
mxu = mean(xu) 
sxu = std(xu)
mxt = mean(xt) 
sxt = std(xt)
mxc = mean(xc) 
sxc = std(xc)

mxua = mean(xua) 
sxua = std(xua)
mxta = mean(xta) 
sxta = std(xta)
mxca = mean(xca) 
sxca = std(xca)
```

### create a list of dicts, and then convert to dataframe

```python
#mydata = [{'subid' : 'B14-111', 'age': 75, 'fdg':1.78},
#          {'subid' : 'B14-112', 'age': 22, 'fdg':1.56},]
mydata = [{'lesion-free' : mxu, 'lesion-free+ack': mxua, 'trans':mxt, 'trans+ack':mxta, 'cis':mxc, 'cis+ack':mxca},]
df = pd.DataFrame(mydata).reset_index()
df[:5]
```

```python
dftrans=df.transpose().drop('index').reset_index()
dftrans.columns=['case','mean']
dftrans["system"]=["cis","cis","lesion-free","lesion-free","trans","trans"]
dftrans["err"]=[sxc,sxca,sxu,sxua,sxt,sxta]

dftrans[:7]
```
