```python
%matplotlib inline

import numpy as np
import pandas as pd
import sys, getopt

from matplotlib import pyplot
from pylab import *
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_color_codes("pastel")
#sns.set(style="white")
sns.set_style("whitegrid")
 
sns.set_context("poster", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.set_style("darkgrid")
```

```python
n_groups =3

lys=(569,685,540)
lys_std=(52,91,101)
ack=(677,460,405)
ack_std=(41,98,61)

lys_text = str(lys[1]) + u"\u00B1" + str(lys_std[1])

index = np.arange(n_groups)
bar_width = 0.15

opacity = 0.4
error_config = {'ecolor': '0.3'}

fig,ax=plt.subplots()
rects1 = plt.bar(index,lys, bar_width,                  
                 color='green',
                 alpha=0.5,
                 yerr=lys_std,
                 error_kw=error_config,
                 label='lys')
 
rects2 = plt.bar(index + bar_width, ack, bar_width,                  
                 color='purple',
                 alpha=0.5,
                 yerr=ack_std,
                 error_kw=error_config,
                 label='ack')
 
for i in np.arange(3): 
    ax.text(index[i]-0.2, lys[i]+ lys_std[i]+10,'%s' % str(lys[i]) + u"\u00B1"+ str(lys_std[i]),fontsize=20)
    ax.text(index[i]+bar_width, ack[i]+ ack_std[i]+10,'%s' % str(ack[i]) + u"\u00B1" + str(ack_std[i]),fontsize=20)

axis([-0.5,3,0,900]) 

plt.ylabel('Contact Surface Area ' r'$(\AA)^2 $',fontsize=25, weight= 'bold')
plt.xticks(index + 2*bar_width, ('lesion-free', 'trans-B[a]P-dG', 'cis-B[a]P-dG'))

plt.legend(loc='center right',numpoints=1, bbox_to_anchor=(1,1),fancybox=True,
 shadow=True, ncol=2)

#plt.tight_layout()
#plt.show()

savefig('NEWcsa_bar_with_error430-460.jpg',figsize=(8, 6), dpi=600, facecolor='w'
, edgecolor='w', orientation='portrait', papertype=None, format=None, transparent=False, bbox_inches=None, pad_inches=0.1, frameon=None)
 
```
